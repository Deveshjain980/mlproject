```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM,GRU,Dense
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
```

```python
url ="https://raw.githubusercontent.com/datasets/finance-vix/master/data/vix-daily.csv"
```

```python
df=pd.read_csv(url)
```

```python
df
```

```python
df['DATE']=pd.to_datetime(df['DATE'])
df=df.sort_values('DATE')
```

```python
data=df[['CLOSE']].values
```

```python
scaler = MinMaxScaler ( feature_range= (0, 1))
data_scaled=scaler.fit_transform(data)
```

```python
def create_sequences(data,seq_length=10):
    X,y=[],[]
    for i in range(len(data) - seq_length):
        X.append(data[i:i+seq_length])
        y.append(data[i+seq_length])
    return np.array(X), np.array(y)
```

```python
seq_length=20
```

```python
X,y=create_sequences(data_scaled,seq_length)
split=int(len(X) *0.8)
X_train,y_train=X[:split], y[:split]
X_test,y_test=X[split:], y[split:]
```

```python
model_lstm=Sequential([
    LSTM(50,return_sequences=True,input_shape=(seq_length,1)),
    LSTM(50,return_sequences=False),
    Dense(25),
    Dense(1)
])
```

```python
model_lstm.compile(optimizer='adam', loss='mse')
history_lstm = model_lstm.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=10, batch_size=32)
```

```python
model_gru = Sequential([
 GRU(50, return_sequences=True, input_shape=(seq_length, 1)),
 GRU(50, return_sequences= False),
 Dense(25),
 Dense(1)
 ])
model_gru.compile(optimizer='adam', loss='mse')
history_gru = model_gru.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=10, batch_size=32)
```

```python
 predicted_lstm = model_lstm.predict(X_test)
 predicted_lstm = scaler.inverse_transform(predicted_lstm)
 predicted_gru = model_gru.predict(X_test)
 predicted_gru = scaler.inverse_transform(predicted_gru)
 actual_prices = scaler.inverse_transform(y_test.reshape(-1, 1))
```

```python
 plt.figure(figsize=(12,6))
 plt.subplot(1,2,1)
 plt.plot(actual_prices, label="Actual", color='blue')
 plt.plot(predicted_lstm, label="LSTM Prediction", color='red')
 plt.title("LSTM Forecasting")
 plt.legend()
 plt.subplot(1,2,2)
 plt.plot(actual_prices, label="Actual", color='blue')
 plt.plot(predicted_gru, label="GRU Prediction", color='green')
 plt.title("GRU Forecasting")
 plt.legend()
 plt.show()
```
